```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

```python
# Class for the cards

class Cards():
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit    
```

```python
#Create a Deck

import random

class Create_A_Deck():
    
    def __init__(self):
        
        self.deck = []
        
        for item in suits:
            for item1 in ranks:
                self.deck.append(Cards(item, item1))
        
        # The reason why we shuffle is that we do not want to get the same "standard" deck everytime!
        random.shuffle(self.deck)
    
    def __str__(self):
        
        for item in self.deck:
            
            print(item)
```

```python
# Do you want to play?

def do_you_want_to_play():
    
    decision = input('Do you want to play? Yes or No: ')
    
    while decision not in ['Yes', 'No']:
        
        decision = input('You can only choose Yes or No! ')
        
    if decision == 'Yes':
        
        return True
    
    else:
        
        return False

# Here, we return a boolean
```

```python
def how_much_do_you_want_to_bet(actual_amount):
    
    amount = input('How much do you want to bet? ')
    
    while amount.isdigit() == False:
        
        amount = input('How much do you want to bet? ')
        
    while int(amount) > actual_amount:
        
        amount = input(f"You don't have enough money. Please enter an amount between 0 and {actual_amount}")
    
    return int(amount) 
```

```python
def initial_cards(Deck):
    
    Player1 = []
    Player2 = []
    a,b,c,d = Deck.deck[0], Deck.deck[1], Deck.deck[2], Deck.deck[3]
    Player1.extend([a, c])
    Player2.extend([b, d])
    
    for item in range(0, 4):
        
        Deck.deck.pop(item)
        
    for item in [a,b,c,d]:
        
        Deck.deck.append(item)
        
    return Player1, Player2, Deck
```

```python
# Take a list and compute the corresponding sum

def sum_ranks_values(listo):
    
    somme = 0 
    
    listo1 = []
    
    for item in listo:
        
        listo1.append(item.rank)
        
    for item in listo:
        
        somme = somme + values[item.rank]
    
    if 'Ace' not in listo1:
        
        return somme
    
    else:
        
        while 'Ace' in listo1:
            
            if somme <= 21:
                
                return somme
            
            else:
                
                somme = somme - 10
                listo1.remove('Ace')
                
    return somme 
```

```python
#Return a boolean

def do_you_want_a_new_card():
    
    decision = input('Do you want a card? Yes or No: ')
    
    while decision not in ['Yes', 'No']:
        
        decision = input('Please, chose Yes or No: ')
        
    if decision == 'Yes':
        
        return True
    
    else:
        
        return False
```

```python
# Game of the bank

def game_bank(my_deck, bank):
    
    total = sum_ranks_values(bank)
    
    listo2 = []
    
    for item in bank:
        
        listo2.append(item.rank)
          
    while total < 17:
        
        popo = my_deck.deck[0]
        bank.append(popo)
        listo2.append(popo.rank)
        my_deck.deck.remove(popo)
        my_deck.deck.append(popo)
        total = sum_ranks_values(bank)
        
        
    return total, bank
    
        
    
```

```python
def do_you_still_want_to_play():
    
    decision = input('Do you still want to play? Yes or No? ')
    
    while decision not in ['Yes', 'No']:
        
         decision = input('Please choose Yes or No ')
            
    if decision == 'Yes':
        
        return True
    
    else:
        
        return False
```
